### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_csv = "Output/weather_data.csv"
df = pd.read_csv(weather_csv)
df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
locations = df[["lat", "lon"]]
humidity = df["Humidity"]
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2.5)

fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.

```python
hotel_df = df.loc[df["Humidity"] < 30]
hotel_df = hotel_df.loc[hotel_df["Max Temp"] < 30]
hotel_df = hotel_df.loc[hotel_df["Max Temp"] > 20]
hotel_df = hotel_df.loc[hotel_df["Cloudiness"] == 0]


len(hotel_df)
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
key = "hotel"

hotelName = []
hotelLat =  []
hotelLon = []
hotelCity = []
hotelCountry = []
hotels_df = pd.DataFrame(columns=['Hotel Name', 'City', 'Country'])

for index, row in hotel_df.iterrows():
    lat = row[2]
    lon = row[3]
    loc_list = (f'{lat}, {lon}')
    city = row[1]
    country = row[8]
    
    params = {
        'location':loc_list,
        'key':g_key,
        'keyword':key,
        'radius':5000
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(base_url, params=params)
    
    hotels = response.json()
    results = hotels['results']
    try:
        hotelName.append(results[0]['name'])
        hotelLat.append(results[0]['geometry']['location']['lat'])
        hotelLon.append(results[0]['geometry']['location']['lng'])
        hotelCity.append(city)
        hotelCountry.append(country)        
        print(results[0]['name'], results[0]['geometry']['location']['lat'], results[0]['geometry']['location']['lng'])
    except IndexError:
        print('No hotel found.')
      
hotels_df["Hotel Name"] = hotelName
hotels_df['Hotel Latitude'] = hotelLat
hotels_df['Hotel Longitude'] = hotelLon
hotels_df['City'] = hotelCity
hotels_df['Country'] = hotelCountry
```
